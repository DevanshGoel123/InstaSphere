```python
# ============================================
# FORMAT DATA FOR TRAINING
# ============================================
# The model needs data in a specific format

def format_example(example):
    return {
        "text": f"""### Instruction:
{example['instruction']}

### Response:
{example['output']}

### End"""
    }

# Apply formatting to all examples
formatted_data = [format_example(ex) for ex in training_data]

# Preview one example
print("📋 Example of formatted data:")
print(formatted_data[0]["text"])
```

```python
from datasets import Dataset

# Convert to HuggingFace Dataset format
dataset = Dataset.from_list(formatted_data)

print(f"✅ Dataset created!")
print(f"Total examples: {len(dataset)}")
print(f"Columns: {dataset.column_names}")
```

```python
from unsloth import FastLanguageModel

print("⏳ Loading model... please wait 2-3 mins")

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/llama-3-8b-bnb-4bit",
    max_seq_length = 2048,
    load_in_4bit = True,        # saves GPU memory ✅
)

print("✅ Base model loaded!")
```

```python
# LoRA makes fine-tuning cheap & fast
model = FastLanguageModel.get_peft_model(
    model,
    r = 16,                     # LoRA rank (higher = more learning)
    lora_alpha = 16,
    lora_dropout = 0,
    target_modules = [
        "q_proj",               # these are parts of
        "k_proj",               # the transformer that
        "v_proj",               # we are fine-tuning
        "o_proj",
    ],
    bias = "none",
    use_gradient_checkpointing = True,
)

print("✅ LoRA adapters added!")
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir = "./finetuned_model",   # where to save
    num_train_epochs = 3,               # go through data 3 times
    per_device_train_batch_size = 2,    # examples per step
    learning_rate = 2e-4,               # how fast to learn
    logging_steps = 5,                  # print progress every 5 steps
    save_steps = 50,                    # save every 50 steps
    warmup_steps = 10,                  # gradual start
    fp16 = True,                        # faster training
    optim = "adamw_8bit",              # memory efficient optimizer
)

print("✅ Training config ready!")
```

```python
from trl import SFTTrainer

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = 2048,
    args = training_args,
)

print("🚀 Starting fine-tuning...")
print("This will take 5-10 minutes on free Colab GPU")
print("You will see loss numbers — lower loss = better learning\n")

trainer.train()

print("\n✅ Fine-tuning COMPLETE!")
```

```python
# Switch model to inference mode (for testing)
FastLanguageModel.for_inference(model)

def ask_model(question):
    prompt = f"""### Instruction:
{question}

### Response:
"""
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

    outputs = model.generate(
        **inputs,
        max_new_tokens = 150,
        temperature = 0.7,
        do_sample = True,
    )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    # Extract only the response part
    response = response.split("### Response:")[-1].strip()
    return response

# Test it!
print("🤖 Testing fine-tuned model...\n")

questions = [
    "What is Python?",
    "What is a loop?",
    "What is a variable?",
]

for q in questions:
    print(f"❓ Question: {q}")
    print(f"🤖 Answer: {ask_model(q)}")
    print("-" * 50)
```
